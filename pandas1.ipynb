```python
market = pd.read_csv('market_fact.csv')
customer = pd.read_csv('cust_dimen.csv')
product = pd.read_csv('prod_dimen.csv')
```

```python
market.head()
```

```python
customer.head()
```

```python
product.head()
```

```python
new_df = pd.merge(market,customer,how ='inner', on ='Cust_id')
```

```python
master_df = pd.merge(new_df,product,how ='inner', on ='Prod_id')
```

```python
master_df.to_csv('master_data.csv',index = False)
```
